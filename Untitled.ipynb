```python
%matplotlib inline
import numpy.random as nr
import matplotlib.pyplot as plt

scale = 2.3

X1 = nr.normal(0, scale, 100)
X2 = nr.normal(0, scale, 100)
X3 = nr.normal(5, scale, 100)
X4 = nr.normal(5, scale, 100)
```

```python
plt.scatter(X1,X2, color='y')
plt.scatter(X3,X4, color='b')

fig = plt.gcf()
ax = fig.gca()

i = 61
ax.set_xlim(-7,7)
ax.set_ylim(-7,7)

plt.axes().set_aspect(1,'datalim')
```

```python
plt.scatter(X1[:50],X2[:50], color='y')
plt.scatter(X3[:50],X4[:50], color='b')


fig = plt.gcf()
ax = fig.gca()

ax.set_xlim(-7,7)
ax.set_ylim(-7,7)



plt.axes().set_aspect(1,'datalim')
```

```python
plt.scatter(X1,X2, color='y')
plt.scatter(X3,X4, color='b')

fig = plt.gcf()
ax = fig.gca()


i = 3
ax.set_xlim(-7,7)
ax.set_ylim(-7,7)

plt.axes().set_aspect(1,'datalim')

c7 = plt.Circle((X1[i],X2[i]),2, color='r',fill=False)
c6 = plt.Circle((X1[i],X2[i]),.2, color='r',fill=True)

ax.add_artist(c6)
ax.add_artist(c7)
```
